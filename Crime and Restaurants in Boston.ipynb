### Convert dates to proper format

```python
restaurants['LicenseAddDtTm'] = pd.to_datetime(restaurants['LicenseAddDtTm'])
crimes['FROMDATE'] = pd.to_datetime(crimes['FROMDATE'])
```

### Convert location to geopy Point object

```python
from geopy.point import Point

# function to convert locations removing parenthesis
def convert_location(loc):
    return Point(loc[1:-1])
```

```python
restaurants.Location = restaurants.Location.apply(convert_location)
crimes.Location = crimes.Location.apply(convert_location)
```

### Filter crimes

```python
# check all types of crimes and incidents
crimes.INCIDENT_TYPE_DESCRIPTION.value_counts().index
```

```python
# list of relevant crimes
relevant = [u'OTHER LARCENY', u'SIMPLE ASSAULT', u'VANDALISM', u'LARCENY FROM MOTOR VEHICLE',
            u'DRUG CHARGES', u'FRAUD', u'RESIDENTIAL BURGLARY', u'AGGRAVATED ASSAULT', u'ROBBERY',
            u'AUTO THEFT', u'Arrest', u'DISORDERLY', u'COMMERCIAL BURGLARY', u'PersMiss', u'FORGERY',
            u'PubDrink', u'TRESPASS', u'WEAPONS CHARGE', u'DEATH INVESTIGATION', u'32GUN', u'Ballist',
            u'STOLEN PROPERTY CHARGES', u'SearchWarr', u'SexReg', u'Restrain',
            u'OPERATING UNDER INFLUENCE', u'PropDam', u'Hazardous', u'CRIMES AGAINST CHILDREN',
            u'VIOLATION OF LIQUOR LAWS', u'Harass', u'PROSTITUTION CHARGES', u'Harbor', u'EMBEZELLMENT',
            u'PRISON', u'Bomb', u'HOMICIDE', u'BurgTools', u'Explos', u'HateCrim', u'ARSON', u'Runaway',
            u'Manslaug', u'BioThreat']
```

```python
# select crime entries
idx = crimes[crimes.INCIDENT_TYPE_DESCRIPTION == relevant.pop()].index
for c in relevant:
    idx = idx.union(crimes[crimes.INCIDENT_TYPE_DESCRIPTION == c].index)
relevant_crimes = crimes.ix[idx]
# check size
relevant_crimes.shape
```

```python
# verify data
relevant_crimes.head()
```

### Attempt to clean location data

```python
zero_point1 = Point('0E-9, 0E-9')
restaurants_no_location = restaurants.Location == zero_point1
# check how many restaurant points to clean
print sum(restaurants_no_location)

zero_point2 = Point('0E-8, 0E-8')
relevant_crimes_no_location = relevant_crimes.Location == zero_point2
# check how many crime points to clean
print sum(relevant_crimes_no_location)
```

```python
# inspect restaurants address
restaurants[restaurants_no_location].head(2)
```

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim()

def fix_location(entry):
    address = str(entry) + ' St Boston MA'
    try:
        location = geolocator.geocode(address, timeout=15)
        if location:
            return Point(location.latitude, location.longitude)
        else:
            address = str(entry) + ' Ave Boston MA'
            location = geolocator.geocode(address, timeout=15)
            if location:
                return Point(location.latitude, location.longitude)
            else:
                address = str(entry) + ' Rd Boston MA'
                location = geolocator.geocode(address, timeout=15)
                if location:
                    return Point(location.latitude, location.longitude)
                else:
                    return zero_point1
    except GeocoderTimedOut:
        return zero_point1
```

```python
# fix restaurants
new_locations = restaurants[restaurants_no_location].Address.apply(fix_location)
restaurants.loc[restaurants_no_location, 'Location'] = new_locations
# check how many restaurant points could not be cleaned
print sum(restaurants.Location == zero_point1)
```

```python
# check some points that could not be cleaned
restaurants[restaurants.Location == zero_point1].head()
```

```python
# save results to avoid calculating again
restaurants[restaurants.Location != zero_point1].to_csv('data/cleaned_restaurants.csv')
```

```python
def fix_location_crime(entry):
    address = str(entry) + ' Boston MA'
    try:
        location = geolocator.geocode(address, timeout=15)
        if location:
            return Point(location.latitude, location.longitude)
        else:
            return zero_point2
    except GeocoderTimedOut:
        return zero_point2
```

```python
# fix crimes
new_locations = relevant_crimes[relevant_crimes_no_location].STREETNAME.apply(fix_location_crime)
relevant_crimes.loc[relevant_crimes_no_location, 'Location'] = new_locations
# check how many crime points could not be cleaned
print sum(relevant_crimes.Location == zero_point2)
```

```python
# check some points that could not be cleaned
relevant_crimes[relevant_crimes.Location == zero_point2].head()
```

```python
# save results to avoid calculating again
relevant_crimes[relevant_crimes.Location != zero_point2].to_csv('data/cleaned_crimes.csv')
```

### Count number of crimes in less than 1/4 mile from restaurant

```python
from geopy.distance import vincenty

# function that returns number of crime occurences near 1/4 mile near a restaurant after it had its license
def get_counts(entry):
    # calculates distance between crime and restaurant
    def get_distance(point1):
        return vincenty(point1, entry.Location).miles
    temp = relevant_crimes.copy()
    temp['distance'] = temp.Location.apply(get_distance)
    l = temp[(temp.FROMDATE >= entry.LicenseAddDtTm) & (temp.distance <= 0.25)].index
    return l.size

restaurants['num_crimes'] = restaurants.apply(get_counts, axis=1)
```

```python
# save results to avoid calculating again
restaurants.to_csv('data/restaurants.csv')
```
